```python
# Clean Data
def clean_data(df):
    df.columns = df.columns.str.strip()
    df = df.drop_duplicates()
    return df

daily = clean_data(daily)
weekly = clean_data(weekly)
monthly = clean_data(monthly)
hourly = clean_data(hourly)
```

```python
# Create Total Sales column
daily['Total_Sales'] = daily[['M01AB','M01AE','N02BA','N02BE','N05B','N05C','R03','R06']].sum(axis=1)

weekly['Total_Sales'] = weekly[['M01AB','M01AE','N02BA','N02BE','N05B','N05C','R03','R06']].sum(axis=1)

monthly['Total_Sales'] = monthly[['M01AB','M01AE','N02BA','N02BE','N05B','N05C','R03','R06']].sum(axis=1)

hourly['Total_Sales'] = hourly[['M01AB','M01AE','N02BA','N02BE','N05B','N05C','R03','R06']].sum(axis=1)
```

```python
# Add Revenue Column

for df in [daily, weekly, monthly, hourly]:
    df['Revenue'] = df['Total_Sales']
```

```python
# Convert Date Columns

daily['datum'] = pd.to_datetime(daily['datum'], errors='coerce')
weekly['datum'] = pd.to_datetime(weekly['datum'], errors='coerce')
monthly['datum'] = pd.to_datetime(monthly['datum'], errors='coerce')
hourly['datum'] = pd.to_datetime(hourly['datum'], errors='coerce')
```

```python
# Save Clean Files

daily.to_csv("D:/clean_daily.csv", index=False)
weekly.to_csv("D:/clean_weekly.csv", index=False)
monthly.to_csv("D:/clean_monthly.csv", index=False)
hourly.to_csv("D:/clean_hourly.csv", index=False)
```
